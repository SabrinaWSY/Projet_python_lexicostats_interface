```python
#实现增加xml标签并转换成xml的代码
import re

#---------------------打开corpus----------------
with open('./corpus(sansEspace/1', 'r') as f:
    corpus = f.read()
    f.close()

words_corpus, ners_corpus = fool.analysis(corpus)
str_resultat = str(ners_corpus)  #将REN抽取结果字符串化
#------------------------------------------------

#====================匹配所需的EN========================
datepat_nr = re.compile(r"\((\d+),\s(\d+),\s'(company|person|location)',\s'(\w+)'\)")
liste_nr = datepat_nr.findall(str_resultat)  #将匹配到的所有元素存入一个变量，此变量属性为列表
#=======================================================

#------------------在文本中插入balise------------------
list_corpus = list(corpus) #将字符串的corpus转换成liste以便于插入标签

n = 0
for i in liste_nr:
    a = int(i[0])
    a = a + n
    b = int(i[1])
    b = b + n
    list_corpus.insert(a,"<"+str(i[2])+">")
    list_corpus.insert(b,"</"+str(i[2])+">")
    n += 2
resultatAvecBalise = "".join(list_corpus)
#--------------------------------------------------
#----------XML化---------------------------
resultXML = "<?xml version=\"1.0\"?>\n"+"<text>\n"+resultatAvecBalise+"</text>"

with open('corpusXML/fool/1.xml', 'w') as a:
    a.write(resultXML)
```

```python
#函数化

#实现增加xml标签并转换成xml的代码
def creatXML(corpusOri,newName):
    import re
    pathCorpusOri = 'corpus(sansEspace/'+ corpusOri
    #---------------------打开corpus----------------
    with open(pathCorpusOri, 'r') as f:
        corpus = f.read()
        f.close()

    words_corpus, ners_corpus = fool.analysis(corpus)
    str_resultat = str(ners_corpus)  #将REN抽取结果字符串化
    #------------------------------------------------

    #====================匹配所需的EN========================
    datepat_nr = re.compile(r"\((\d+),\s(\d+),\s'(company|person|location)',\s'(\w+)'\)")
    liste_nr = datepat_nr.findall(str_resultat)  #将匹配到的所有元素存入一个变量，此变量属性为列表
    #=======================================================

    #------------------在文本中插入balise------------------
    list_corpus = list(corpus) #将字符串的corpus转换成liste以便于插入标签

    n = 0
    for i in liste_nr:
        a = int(i[0])
        a = a + n
        b = int(i[1])
        b = b + n
        list_corpus.insert(a,"<"+str(i[2])+">")
        list_corpus.insert(b,"</"+str(i[2])+">")
        n += 2
    resultatAvecBalise = "".join(list_corpus)
    #--------------------------------------------------
    #----------XML化---------------------------
    resultXML = "<?xml version=\"1.0\"?>\n"+"<text>\n"+resultatAvecBalise+"</text>"

    pathCorpusNew = 'corpusXML/fool/' + newName + '.xml'
    with open(pathCorpusNew, 'w') as a:
        a.write(resultXML)
```
